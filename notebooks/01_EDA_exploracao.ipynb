## 📊 4. Análise da Variável Target (Diagnóstico)

```python
print("🎯 DISTRIBUIÇÃO DO DIAGNÓSTICO")
print("="*80)

# Contagem
target_counts = df['diagnosis'].value_counts()
target_pct = df['diagnosis'].value_counts(normalize=True) * 100

print(f"\n📊 Contagem absoluta:")
for diagnosis, count in target_counts.items():
    print(f"   {diagnosis}: {count} ({target_pct[diagnosis]:.2f}%)")

# Verificar balanceamento
diff = abs(target_pct.iloc[0] - target_pct.iloc[1])
if diff < 10:
    status = "✅ BALANCEADO"
elif diff < 20:
    status = "⚠️ LEVEMENTE DESBALANCEADO"
else:
    status = "❌ DESBALANCEADO"

print(f"\n{status}")
print(f"Diferença entre classes: {diff:.2f}%")
```

```python
# Visualização da distribuição do target
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico de barras
colors = ['#FF6B6B', '#4ECDC4']  # Vermelho para Maligno, Verde-azulado para Benigno
target_counts.plot(kind='bar', ax=axes[0], color=colors, alpha=0.8, edgecolor='black')
axes[0].set_title('Distribuição dos Diagnósticos', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Diagnóstico', fontsize=12)
axes[0].set_ylabel('Contagem', fontsize=12)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
axes[0].grid(axis='y', alpha=0.3)

# Adicionar valores nas barras
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v, f'{v}\n({target_pct.values[i]:.1f}%)', 
                ha='center', va='bottom', fontweight='bold')

# Gráfico de pizza
axes[1].pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%',
           colors=colors, startangle=90, explode=(0.05, 0.05),
           shadow=True, textprops={'fontsize': 11, 'fontweight': 'bold'})
axes[1].set_title('Proporção dos Diagnósticos', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../reports/figures/01_distribuicao_target.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/01_distribuicao_target.png")
```

## 🔍 5. Análise de Valores Ausentes

```python
print("❓ ANÁLISE DE VALORES AUSENTES")
print("="*80)

# Verificar missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Coluna': missing.index,
    'Missing': missing.values,
    'Percentual': missing_pct.values
})

missing_df = missing_df[missing_df['Missing'] > 0].sort_values('Percentual', ascending=False)

if len(missing_df) == 0:
    print("\n✅ EXCELENTE! Nenhum valor ausente encontrado!")
    print("   O dataset está completo e pronto para análise.")
else:
    print(f"\n⚠️ {len(missing_df)} colunas com valores ausentes:")
    print(missing_df.to_string(index=False))
```

## 📊 6. Estatísticas Descritivas

```python
print("📈 ESTATÍSTICAS DESCRITIVAS - FEATURES NUMÉRICAS")
print("="*80)

# Selecionar apenas colunas numéricas (exceto target)
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop('target')

df[numeric_cols].describe().T
```

```python
# Estatísticas adicionais
print("📊 ESTATÍSTICAS ADICIONAIS")
print("="*80)

stats_df = pd.DataFrame({
    'Feature': numeric_cols,
    'Média': df[numeric_cols].mean(),
    'Mediana': df[numeric_cols].median(),
    'Desvio Padrão': df[numeric_cols].std(),
    'Variância': df[numeric_cols].var(),
    'Assimetria': df[numeric_cols].skew(),
    'Curtose': df[numeric_cols].kurtosis()
})

print("\n5 Features com maior variabilidade (desvio padrão):")
print(stats_df.nlargest(5, 'Desvio Padrão')[['Feature', 'Desvio Padrão']])

print("\n5 Features com maior assimetria:")
print(stats_df.nlargest(5, 'Assimetria')[['Feature', 'Assimetria']])
```

## 📊 7. Distribuição das Features Principais

```python
# Selecionar features 'mean' (média)
mean_features = [col for col in df.columns if 'mean' in col]

print(f"📊 Analisando {len(mean_features)} features principais (mean):")
print(mean_features)
```

```python
# Histogramas das features principais
n_cols = 3
n_rows = (len(mean_features) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
axes = axes.flatten()

for idx, col in enumerate(mean_features):
    axes[idx].hist(df[col], bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    axes[idx].axvline(df[col].mean(), color='red', linestyle='--', 
                     linewidth=2, label=f'Média: {df[col].mean():.2f}')
    axes[idx].axvline(df[col].median(), color='green', linestyle='--', 
                     linewidth=2, label=f'Mediana: {df[col].median():.2f}')
    axes[idx].set_title(col, fontweight='bold')
    axes[idx].set_xlabel('Valor')
    axes[idx].set_ylabel('Frequência')
    axes[idx].legend(fontsize=8)
    axes[idx].grid(alpha=0.3)

# Remover subplots extras
for idx in range(len(mean_features), len(axes)):
    fig.delaxes(axes[idx])

plt.suptitle('Distribuição das Features Principais (Mean)', 
             fontsize=16, fontweight='bold', y=1.001)
plt.tight_layout()
plt.savefig('../reports/figures/02_distribuicoes_features.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/02_distribuicoes_features.png")
```

## 📦 8. Análise de Outliers com Boxplots

```python
# Boxplots para identificar outliers
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
axes = axes.flatten()

for idx, col in enumerate(mean_features):
    bp = axes[idx].boxplot(df[col].dropna(), patch_artist=True, 
                           boxprops=dict(facecolor='lightblue', alpha=0.7),
                           medianprops=dict(color='red', linewidth=2),
                           whiskerprops=dict(color='blue', linewidth=1.5),
                           capprops=dict(color='blue', linewidth=1.5))
    axes[idx].set_title(col, fontweight='bold')
    axes[idx].set_ylabel('Valor')
    axes[idx].grid(alpha=0.3)
    
    # Calcular e mostrar estatísticas de outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    axes[idx].text(0.5, 0.95, f'Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)',
                  transform=axes[idx].transAxes, ha='center', va='top',
                  bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                  fontsize=8)

# Remover subplots extras
for idx in range(len(mean_features), len(axes)):
    fig.delaxes(axes[idx])

plt.suptitle('Análise de Outliers - Boxplots', fontsize=16, fontweight='bold', y=1.001)
plt.tight_layout()
plt.savefig('../reports/figures/03_boxplots_outliers.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/03_boxplots_outliers.png")
```

## 🔗 9. Matriz de Correlação

```python
# Calcular matriz de correlação (apenas features mean)
correlation_matrix = df[mean_features].corr()

# Visualização
plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.2f', 
            cmap='coolwarm', center=0, square=True, linewidths=1,
            cbar_kws={"shrink": 0.8}, vmin=-1, vmax=1)
plt.title('Matriz de Correlação - Features Principais (Mean)', 
         fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../reports/figures/04_matriz_correlacao.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/04_matriz_correlacao.png")
```

```python
# Identificar correlações fortes (|r| > 0.7)
print("🔗 CORRELAÇÕES FORTES (|r| > 0.7)")
print("="*80)

strong_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            strong_corr.append({
                'Feature 1': correlation_matrix.columns[i],
                'Feature 2': correlation_matrix.columns[j],
                'Correlação': correlation_matrix.iloc[i, j]
            })

strong_corr_df = pd.DataFrame(strong_corr).sort_values('Correlação', 
                                                        ascending=False, 
                                                        key=abs)
print(strong_corr_df.to_string(index=False))
```

## 🎯 10. Correlação com Target

```python
# Calcular correlação de todas features com target
target_corr = df[numeric_cols].corrwith(df['target']).sort_values(ascending=False)

print("🎯 CORRELAÇÃO DAS FEATURES COM TARGET (Diagnóstico)")
print("="*80)
print("\n📊 Top 10 features mais correlacionadas (positivas):")
print(target_corr.head(10))
print("\n📊 Top 10 features mais correlacionadas (negativas):")
print(target_corr.tail(10))
```

```python
# Visualização da correlação com target
plt.figure(figsize=(10, 12))
target_corr_sorted = target_corr.abs().sort_values(ascending=True)
colors = ['green' if x > 0 else 'red' for x in target_corr[target_corr_sorted.index]]

plt.barh(range(len(target_corr_sorted)), target_corr[target_corr_sorted.index], color=colors, alpha=0.7)
plt.yticks(range(len(target_corr_sorted)), target_corr_sorted.index, fontsize=8)
plt.xlabel('Correlação com Target', fontsize=12, fontweight='bold')
plt.title('Correlação das Features com Diagnóstico\n(Verde: Positiva | Vermelho: Negativa)', 
         fontsize=14, fontweight='bold')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('../reports/figures/05_correlacao_target.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/05_correlacao_target.png")
```

## 📊 11. Distribuições por Diagnóstico

```python
# Selecionar top 6 features mais correlacionadas
top_features = target_corr.abs().nlargest(6).index

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, feature in enumerate(top_features):
    for diagnosis in df['diagnosis'].unique():
        subset = df[df['diagnosis'] == diagnosis][feature]
        color = '#FF6B6B' if diagnosis == 'Maligno' else '#4ECDC4'
        axes[idx].hist(subset, bins=25, alpha=0.6, label=diagnosis, 
                      color=color, edgecolor='black')
    
    axes[idx].set_title(feature, fontweight='bold', fontsize=11)
    axes[idx].set_xlabel('Valor', fontsize=10)
    axes[idx].set_ylabel('Frequência', fontsize=10)
    axes[idx].legend(fontsize=9)
    axes[idx].grid(alpha=0.3)

plt.suptitle('Distribuição das Top 6 Features por Diagnóstico', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/06_distribuicao_por_diagnostico.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/06_distribuicao_por_diagnostico.png")
```

## 📈 12. Pairplot das Features Principais

```python
# Selecionar top 4 features para pairplot (para não ficar muito pesado)
top_4_features = target_corr.abs().nlargest(4).index.tolist()
top_4_features.append('diagnosis')

print(f"Criando pairplot para: {top_4_features}")

# Criar pairplot
pairplot_df = df[top_4_features]
g = sns.pairplot(pairplot_df, hue='diagnosis', 
                palette={'Maligno': '#FF6B6B', 'Benigno': '#4ECDC4'},
                diag_kind='kde', plot_kws={'alpha': 0.6},
                height=2.5)
g.fig.suptitle('Pairplot - Top 4 Features Mais Correlacionadas', 
              fontsize=14, fontweight='bold', y=1.01)
plt.savefig('../reports/figures/07_pairplot.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Gráfico salvo em: reports/figures/07_pairplot.png")
```
